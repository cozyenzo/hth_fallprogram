```python
### Question 1) What are the last 37 Pokemon in the PokeDex?
import pandas
pokemon_data = pandas.read_csv('PokemonData.csv')
newest_25_pokemon = pokemon_data.tail(37)
newest_25_pokemon
```

```python
# Question) How many pokemon do not have a second typing?

import pandas
pokemon_data = pandas.read_csv('PokemonData.csv')
pokemon_type_missing = pokemon_data['Type2'].isnull().sum()
pokemon_type_missing
```

```python
# Question 3) What is the highest attack stat? 

import pandas
pokemon_data = pandas.read_csv('PokemonData.csv')

maxattack = pokemon_data['Attack'].max()
maxattack
```

```python
# Question 3) What is the minimum attack stat?

import pandas
pokemon_data = pandas.read_csv('PokemonData.csv')

minattack = pokemon_data['Attack'].min()
minattack
```
